```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import *
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate, LeaveOneOut
import seaborn as sns
%matplotlib inline
```

```python
sns.set(style="white", palette="muted", color_codes=True)
```

```python
X, Y = make_regression(n_features=3, bias=5.0, noise = 7.0, n_samples = 1000)
model = LinearRegression()
```

## K-Folds

```python
cross_val_score(model, X, Y, cv=10, scoring="mean_squared_error")
```

```python
cross_validate(model, X, Y, cv=10, scoring="mean_squared_error", return_train_score=False)
```

## Leave One Out

```python
cross_val_score(model, X, Y, cv=LeaveOneOut(), scoring="mean_squared_error")
```

```python
cross_validate(model, X, Y, cv=LeaveOneOut(), scoring="mean_squared_error", return_train_score=False)
```
