```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import *
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
X, Y = make_regression(n_features=1, bias=3, noise = 7.0, n_samples=1000)
```

```python
plt.scatter(X,Y)
```

```python
x_train, x_test = np.split(X, [int(np.floor(len(X)*0.8))])
y_train, y_test = np.split(Y, [int(np.floor(len(Y)*0.8))])
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predicted = model.predict(x_train)
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_predicted, color='red')
```

```python
print("MSE=%s" % mean_squared_error(y_train, y_predicted))
```
