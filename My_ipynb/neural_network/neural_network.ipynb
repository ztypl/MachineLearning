```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import *
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
%matplotlib inline

#sns.set(style="white", palette="muted", color_codes=True)
```

```python
X, y = make_moons(n_samples=5000, noise = 0.2, random_state=0)
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Spectral)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=0)
model = MLPClassifier(hidden_layer_sizes=(10,10), random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```

```python
h = 0.02
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5  
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5  
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
plt.contourf(xx, yy, Z.reshape(xx.shape), cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(x_test[:,0], x_test[:,1], c=y_test, cmap=plt.cm.Spectral)
```
